```python
# How many people have Credit Card that Expires in 2020

data['CC Exp Date']
```

```python
def fun():
    count=0
    for date in data['CC Exp Date']:
        if date.split('/') [1] == '20':
            count=count+1                               #   The First Method 
        print(count)
```

```python
fun()
```

```python
data[['Exp month' , 'Exp year']] = data['CC Exp Date'].str.split('/', expand=True) 
                                                                                               # The Second Method
```

```python
len(data[data['Exp year'] =='20'])  # Best Method     # The Second Method
```

```python
len(data[data['CC Exp Date'].apply(lambda x:x[3:]=='20')])  # The Best and the Third Method 
```
